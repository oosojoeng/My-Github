```python
def add_sub(a, b):
  result_add = a + b
  result_sub = a - b
  return result_add, result_sub
```

```python
a = 10
b = 20
c = add_sub(a,b)
c
```

```python
d, e = add_sub(100, 200)
print(d)
print(e)
```
